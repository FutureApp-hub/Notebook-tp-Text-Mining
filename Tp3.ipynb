## Travail pratique no 3: Création d'un nuage de mots
Dans cette partie nous donnons fonctions essentielles de word_cloud.
Un nuage de mots est une image composée de mots proportionnel à leur taille.
La taille d'un mot dans l'image montre à quel point il est important.
Les nuages de mots sont en général utilisés pour Résumer un texte long, pour visualiser (graphiquement) le contenu d'un texte, etc.

### Module nécessaire

```python
#Module qui va créer notre image (nuage) à partir du texte
from wordcloud import WordCloud
#Module qui nous permet par la suite d'afficher notre image
from matplotlib import pyplot as plt
```

### Application

```python
text = """Dans cette partie nous donnons fonctions essentielles de word_cloud. Un nuage de mots est une image composée de mots proportionnel à leur taille.
 La taille d'un mot dans l'image montre à quel point il est important. Les nuages de mots sont en général utilisés pour Résumer un texte long, pour 
 visualiser (graphiquement) le contenu d'un texte, etc."""
```

```python
wordcloud_instance = WordCloud(width = 800, height = 800, background_color ='black', stopwords=None, min_font_size = 10).generate(text) 
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud_instance) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
